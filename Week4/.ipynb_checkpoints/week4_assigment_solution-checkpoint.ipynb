```python
import PyPDF2
import nltk
import re
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem import WordNetLemmatizer
from gensim.models import Word2Vec
```

```python
pdf_file_name = "Arunachal_Pradesh_britannia.pdf" 
```

```python
def extract_text_from_pdf(pdf_path):
    try:
        with open(pdf_path, 'rb') as file:
            reader = PyPDF2.PdfReader(file)
            text = ""
            for page in reader.pages:
                text += page.extract_text()
            return text
    except Exception as e:
        print(f"An error occurred: {e}")
        return None
```

```python
text = extract_text_from_pdf(pdf_file_name)

if text:
    print("Extracted Text:")
    print(text)
```

## Pre-Processing steps

```python
lower_case_text = text.lower()
```

```python
# print(lower_case_text)
```

```python
en_stopwords = stopwords.words('english')
```

```python
# print(en_stopwords)
```

```python
text_removed_stopwords = ' '.join([word for word in lower_case_text.split() if word not in en_stopwords])
```

```python
print(text_removed_stopwords)
```

```python
punctuation_pattern_to_find = r"[^\w\s]"
```

```python
text_wo_punctuation = re.sub(punctuation_pattern_to_find, "", text_removed_stopwords)
```

```python
text_tokenized_by_word = word_tokenize(text_wo_punctuation)
```

```python
print(text_tokenized_by_word)
```

```python
# unique_word_set = set(text_tokenized_by_word)
```

### Lemmatized words in the dataset

```python
lemmatizer = WordNetLemmatizer()
```

```python
for word in text_tokenized_by_word:
    print(word , ":" , lemmatizer.lemmatize(word))
```

## Generating Embeddings for a text dataset

```python
# Train a Word2Vec model
model = Word2Vec(
    sentences=text_tokenized_by_word,  # Input data (list of tokenized sentences)
    vector_size=100,          # Dimensionality of the word vectors
    window=5,                 # Context window size
    min_count=1,              # Minimum word frequency to consider
    workers=4,                # Number of worker threads to train the model
    sg=1                      # Skip-gram model (set to 0 for CBOW)
)

# Get the vocabulary of the trained model
vocab = model.wv.key_to_index
```

```python
Generate embeddings for all sentences in the dataset
sentence_embeddings = []
for sentence in text_tokenized_by_word:
    # Average word embeddings for the sentence
    embedding = sum(model.wv[word] for word in sentence if word in vocab) / len(sentence)
    sentence_embeddings.append(embedding)

print("\nSentence embeddings:")
for i, embedding in enumerate(sentence_embeddings):
    print(f"Sentence {i+1}: {embedding}")
```
